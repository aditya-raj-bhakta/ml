```python
class StochasticGDregrssor:
    def __init__(self,learning_rate,epoches):
        self.lr=learning_rate
        self.epoches=epoches
        self.coeff=None
    def fit(self,x,y):
        self.coeff=np.ones(x.shape[1])
        self.coeff=np.insert(self.coeff,0,0).reshape(1,-1)
        x=np.insert(x,0,1,axis=1)
        for i in range(self.epoches):
            for j in range(x.shape[0]):
                idx=np.random.randint(0,x.shape[0])
                y_pred=np.dot(self.coeff,x[idx].reshape(-1,1))
                self.coeff=self.coeff-self.lr*(-2*(np.dot((y[idx]-y_pred),(x[idx].reshape(1,-1)))))
        print(self.coeff)
```

```python
class GDregrssor:
    def __init__(self,learning_rate,epoches):
        self.lr=learning_rate
        self.epoches=epoches
        self.coeff=None
    def fit(self,x,y):
        self.coeff=np.ones(x.shape[1])
        self.coeff=np.insert(self.coeff,0,0).reshape(-1,1)
        x=np.insert(x, 0, 1, axis=1)
        y=y.reshape(-1,1)
        for i in range(self.epoches):
            y_pred=np.dot(x,self.coeff)
            gradient=np.dot((y-y_pred).T,x)/x.shape[0]*-2
            self.coeff=self.coeff-(self.lr*gradient).T
        print(self.coeff)
```

```python
model1=StochasticGDregrssor(0.0001,10000)
model2=GDregrssor(0.0001,10000)
model3=LinearRegression()
model1.fit(x,y)
model2.fit(x,y)
model3.fit(x,y)
print(model3.coef_)
```

```python
class GDRegressor:

    def __init__(self, learning_rate=0.01, epochs=1000):
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, x, y):

        x = np.insert(x, 0, 1, axis=1)
        y = y.reshape(-1,1)

        self.coeff = np.zeros((x.shape[1],1))

        for _ in range(self.epochs):

            y_pred = np.dot(x, self.coeff)

            gradient = -(2/x.shape[0]) * np.dot(x.T, (y-y_pred))

            self.coeff -= self.lr * gradient

        return self
```

```python
model=GDRegressor(learning_rate=0.0001, epochs=10000)
model.fit(x,y)
print(model.coeff)
```
