```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder,StandardScaler

np.set_printoptions(suppress=True)
```

```python
df=pd.read_csv("logistic-regression/Social_Network_Ads.csv")
df["Gender"]
```

```python
encoder=OneHotEncoder(sparse_output=False)
encoded=encoder.fit_transform(df[["Gender"]])
df_ecoded=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(),index=df.index)
df=pd.concat([df,df_ecoded],axis=1)
df=df.drop("Gender",axis=1)
```

```python
class LogisticRegression:
    def __init__(self,lr,epochs):
        self.coef_=None
        self.lr=lr
        self.epochs=epochs
    def fit(self,x,y):
        self.coef_=np.ones(x.shape[1])
        x=np.insert(x,0,1,axis=1)
        self.coef_=np.insert(self.coef_,0,0)
        for i in range(self.epochs):
            idx=np.random.randint(0,x.shape[0])
            y_pred=1 if np.dot(x[idx],self.coef_)>0 else 0
            print(y_pred,y[idx])
            self.coef_=self.coef_+self.lr*(y[idx]-y_pred)*x[idx]
    def predict(self,x):
        x=np.insert(x,0,1,axis=1)
        return  np.dot(x,self.coef_)
        
```

```python
model=LogisticRegression(0.01,100)
x=df[['Age','EstimatedSalary','Gender_Male',"Gender_Female"]]
y=df['Purchased']
```

```python
model.fit(x,y )
```
